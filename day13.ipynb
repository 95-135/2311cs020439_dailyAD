```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
df=sns.load_dataset('taxis')
```

```python
df
```

```python
sns.barplot(x='color',y='total',data=df,color='red')
plt.show()
plt.tight_layout()
```

```python
sns.histplot(x=df['fare'],palette=['blue','red'],hue=df['payment'])#palette for more than one color we can  use
plt.show()
```

```python
sns.violinplot(x=df['fare'],y=df['payment'])
```

```python
sns.boxplot(x=df['fare'],y=df['payment'],palette=['green','yellow'])
```
